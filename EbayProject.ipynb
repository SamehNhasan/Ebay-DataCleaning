##### Filling missing values <a id="Filling_missing_values"></a>

```python
## Create function to extract brand from the Product_Description column
def extracted_brand(desc):
    for brand in ['Dell', 'Lenovo', 'HP']:
        if brand in str(desc):
            return brand
    np.nan
## Use Apply function to replace null values in brand with the brand in Product_Description column
ebay['Brand'] = ebay['Brand'].fillna(ebay['Product_Description'].apply(extracted_brand))   
```

#### Categorize Processor column and adding new column with categories <a id="categorize-processor-column-and-adding-new-column-with-categories"></a>

```python
ebay['Processor'] = ebay['Processor'].str.lower()
def categorize(name):
    if 'core i3' in name:
        return 'Core i3'
    elif 'core i5' in name:
        return 'Core i5'
    elif 'core i7' in name:
        return 'Core i7'
    elif 'core i9' in name:
        return 'Core i9'
    elif 'ryzen 3' in name:
        return 'Ryzen 3'
    elif 'ryzen 5' in name:
        return 'Ryzen 5'
    elif 'ryzen 7' in name:
        return 'Ryzen 7'
    elif 'ryzen 9' in name:
        return 'Ryzen 9'
    elif 'pentium' in name:
        return 'Pentium'
    elif 'celeron' in name:
        return 'Celeron'
    else:
        return 'Other'
## Use Apply function to replace null values in GPU with the brand in GPU_Type column
ebay['Category'] = ebay['Processor'].apply(categorize)
```
